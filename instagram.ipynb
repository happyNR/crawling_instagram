# 셀레니움을 이용한 인스타그램 크롤링

```python
from bs4 import BeautifulSoup
import time
import requests
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
```

```python
keyword = "크리스마스"
url = "https://www.instagram.com/explore/tags/{}/".format(keyword)

driver = webdriver.Chrome('C:/Users/snari/Downloads/chromedriver_win32/chromedriver.exe')
driver.implicitly_wait(2)
driver.get(url)
```

```python
reallink = []

while len(reallink) < 33:
    pageString = driver.page_source
    soup = BeautifulSoup(pageString, "lxml")

    for link1 in soup.find_all(name="div",attrs={"class":"Nnq7C weEfm"}):
        title = link1.select('a')[0] 
        real = title.attrs['href']
        reallink.append('https://www.instagram.com' + real) 
        title = link1.select('a')[1] 
        real = title.attrs['href']
        reallink.append('https://www.instagram.com' + real) 
        title = link1.select('a')[2] 
        real = title.attrs['href']
        reallink.append('https://www.instagram.com' + real)

reallinknum = len(reallink)
print("총"+str(reallinknum)+"개의 URL 수집")
#공유복사 pyperclip.paste(붙여넣기)
```
